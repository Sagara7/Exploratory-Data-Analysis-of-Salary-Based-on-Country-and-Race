# Data Cleaning

Setelah sebelumnya melihat data yang akan digunakan, selanjutnya dilakukan pembersihan terhadap bagian yang tidak diperlukan pada data tersebut. Pertama - tama, pada kolom yang berisikan index memiliki nama 'unnamed: 0'. Kolom tersebut akan dihapus sehingga hanya ditandakan sebagai index yang menunjukkan jumlah data seseorang yang digunakan untuk penelitian.

```python
# Menghapus kolom yang bernamakan 'Unnamed: 0'
df.drop('Unnamed: 0', axis = 1, inplace = True)
```

Setelah menghapus kolom tersebut, selanjutnya akan dilakukan pengecekan terhadap missing value yang ada pada data. Missing value tersebut akan dikeluarkan dari tabel sehingga tidak akan mengganggu analisa yang akan dilakukan nanti.

```python
# Menampilkan jumlah missing value dari masing - masing kolom
df.isnull().sum()
```

```python
# Menghapus setiap kolom yang merupakan missing value
df = df.dropna()
df
```

Penghapusan missing value selesai sehingga baris yang sebelumnya berjumlah 6704 baris berkurang menjadi 6698 baris. Setelah itu dilakukan pengecekan jumlah nilai unik dari masing - masing kolom untuk mengetahui apakah ada yang tidak sesuai dengan yang dibutuhkan.

```python
# Melihat jumlah nilai unik dari masing - masing kolom
df.nunique()
```

Dapat dilihat dari kolom gender terdapat tiga buah nilai unik. Karena data yang ingin diperoleh hanya membutuhkan dua tipe gender, maka dari itu harus dilihat dulu nilai unik dari kolom gender.

```python
# Memperlihatkan nilai unik dari kolom 'Gender'
df['Gender'].unique()
```

Dengan adanya nilai unik dengan nama 'Other', setiap kolom yang berisikan nilai tersebut perlu dihapus.

```python
# Melihat jumlah value dari masing - masing nilai unik pada kolom 'Gender'
df['Gender'].value_counts()
```

```python
# Menghapus 'Other' pada kolom 'Gender'
df = df[~(df.Gender == 'Other')]
```

Pada kolom Education Level, terdapat 7 nilai unik yaitu High School, phD, PhD, Bachelor's, Bachelor's Degree, Master's, dan Master's Degree. Disini nilai unik akan dikurangi menjadi 4 saja karena ada beberapa data yang memiliki arti sama dan disatukan sehingga memiliki nama yang sama.

```python
# Mengganti phD menjadi PhD, Bachelor's menjadi Bachelor's Degree, dan Master's menjadi Master's Degree.
df["Education Level"].replace("phD", "PhD", inplace = True)
df["Education Level"].replace("Bachelor's", "Bachelor's Degree", inplace = True)
df["Education Level"].replace("Master's", "Master's Degree", inplace = True)
```

Karena sebelumnya sudah dilakukan penghapusan missing value, kolom index jadi tidak berurutan, maka dari itu akan dibuat index baru agar kembali berurutan.

```python
# Melakukan reset pada kolom index
df = df.reset_index()
```

```python
# Melakukan drop pada kolom index
df.drop("index", axis = 1, inplace = True)
```

```python
# Melihat apakah index sesuai dengan jumlah baris
df
```

# Analisis dan Perhitungan

### 1. Mencari nilai mean, median, dan modus dari kolom gaji.

```python
# Mencari nilai rata - rata dari kolom gaji
rata_gaji = df['Salary'].mean()
rata_gaji

# Mencari nilai median dari kolom gaji
median_gaji = df['Salary'].median()
median_gaji

# Mencari nilai modus dari kolom gaji
modus_gaji = df['Salary'].mode().values[0]
modus_gaji

print(f'Rata - rata gaji pekerja adalah: {rata_gaji}')
print(f'Median gaji pekerja adalah: {median_gaji}')
print(f'Modus gaji pekerja adalah: {modus_gaji}')
```

Dari data yang dimiliki, gaji rata - rata dari para pekerja mendekati nilai median gaji dari seluruh pekerja. Serta paling banyak pekerja memiliki gaji sebesar 140.000. Dari sini kita akan analisa faktor apa saja yang memiliki pengaruh terhadap gaji seseorang.

### 2. Apakah pengalaman bekerja memiliki pengaruh terhadap gaji?

Setelah mengetahui rata - rata, median, dan modus dari gaji para pekerja selanjutnya ingin diketahui apakah gaji seorang pekerja memiliki pengaruh yang diberikan oleh pengalaman bekerja dari si pekerja tersebut. Disini digunakan perhitungan statistik inferential untuk mengetahui hubungan antara kedua aspek tersebut.

```python
# Mencari nilai skew
skewness_gaji = df['Salary'].skew()
skewness_exp = df['Years of Experience'].skew()

print(f'Nilai skewness kolom Salary: {skewness_gaji}')
print(f'Nilai skewness kolom Years of Experience: {skewness_exp}')
```

Distribusi data masuk ke dalam kategori normal karena nilai skewness berada pada nilai dibawah 1. Selanjutnya dilakukan uji korelasi untuk mengetahui hubungan kedua variabel.

Uji Korelasi

Asumsi:

H0 = Tidak ada korelasi linier antara gaji dan pengalaman bekerja.

H1 = Ada korelasi linier antara gaji dan pengalaman bekerja.

```python
# Mencari korelasi antara kedua variabel dengan menggunakan metode Pearson
correlation = df['Salary'].corr(df['Years of Experience'])

print(f"Korelasi Pearson antara gaji dan pengalaman bekerja: {correlation}")

if correlation == 0:
    print("Tidak ada korelasi antara gaji and pengalaman bekerja.")
elif correlation > 0:
    print("Terdapat korelasi linear positif antara gaji and pengalaman bekerja.")
else:
    print("Terdapat korelasi linear negatif antara gaji and pengalaman bekerja.")
```

Kesimpulan dari hubungan antara pengalaman bekerja dengan gaji memiliki hubungan yang sangat kuat karena nilai korelasi lebih besar dari 0.8. Semakin lama seseorang bekerja maka gaji orang tersebut akan semakin besar.

```python
plt.title("Hubungan Antara Pengalaman dengan Gaji")
sns.regplot(data=df,x="Years of Experience",y="Salary")
```

Untuk pembuktian yang lebih kuat, dibuat grafik dalam bentuk scatter. Dengan memanfaatkan metode regresi linier, garis regresi condong ke arah atas. Maka dari itu, semakin banyak pengalaman bekerja dari seseorang maka semakin tinggi juga gaji yang dimiliki seseorang tersebut. Namun, gaji seseorang yang memiliki pengalaman lebih dari 20 tahun cenderung berhenti naik dan berada pada nilai sekitar 200.000.

### 3. Keturunan apa yang paling banyak bekerja?

Dari data yang digunakan, akan dicari keturunan ras mana yang paling banyak bekerja.

```python
fig,ax1 = plt.subplots(figsize=(8,8))

df['Race'].value_counts().plot(kind='pie', autopct='%.2f%%',ax=ax1)
plt.show()
```

Dari pie chart yang sudah dibuat, keturunan ras yang paling banyak adalah 'White', sedangkan yang terbanyak kedua adalah 'Asian'. Selanjutnya akan ditinjau apakah terdapat pengaruh dari keturunan ras terhadap gaji yang mereka miliki.

### 4. Apakah kelompok ras memiliki pengaruh terhadap gaji?

Disini akan dibuat grafik berbentuk bar plot antara keturunan ras dengan gaji dari setiap pegawai sesuai dengan keturunan ras nya. Dengan adanya grafik, dapat dilihat perbedaan pendapatan setiap ras dari data yang digunakan.

```python
# Membuat bar plot
df.groupby('Race')['Salary'].mean().plot(kind = 'bar', figsize = (12,6))
plt.grid(axis = 'y')
```

Setelah dibandingkan rata - rata gaji setiap keturunan ras, tidak terdapat perbedaan signifikan antar setiap keturunan ras. Disini menandakan bahwa pendapatan setiap orang tidak dipengaruhi oleh keturunan ras. Selanjutnya akan dilihat apakah gaji seseorang dipengaruhi oleh riwayat pendidikan atau tidak.

### 5. Apakah riwayat pendidikan berpengaruh terhadap gaji?

```python
# Membuat bar plot dengan menggunakan seaborn
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(x="Education Level", y="Salary", data=df, errorbar=None)
plt.xticks(rotation=45)
plt.xlabel("Education Level")
plt.ylabel("Salary")
plt.title("Average Salary by Education Level")

plt.tight_layout()
plt.show()
```

Dari grafik yang sudah dibuat, pekerja dengan gelar PhD memilki rata - rata gaji paling tinggi, sedangkan lulusan SMA memiliki pendapatan terendah. Hal ini menandakan bahwa tingkat pendidikan memiliki pengaruh yang cukup signifikan terhadap pendapatan seseorang. 

### 6. Jabatan apa yang memiliki gaji paling tinggi?

```python
# Melakukan groupby antara jabatan dengan pendapatan
gaji_tertinggi = df.groupby(["Job Title"])["Salary"].mean().sort_values(ascending = False)
```

```python
# Membuat grafik untuk melihat gaji tertinggi dari masing - masing jabatan
plt.figure(figsize = (10,18))                       # Ukuran plot, 10 inci horizontal dan 18 inci vertikal
plt.subplot(2,1,1)                                  # Membuat subplot berisikan 2 baris dan 1 kolom
ax = sns.barplot(y = gaji_tertinggi[:10].index,     # Pembuatan barplot menggunakan seaborn, serta pengambilan 25 baris pertama
                 x = gaji_tertinggi[:10].values,    
                 palette = 'vlag', width = .85,
                 label = 'Gaji Tertinggi')
ax.set_xlabel('Gaji', fontdict = {'family': 'monospace',        # Setting font, ukuran font, dan warna font pada axis x
                                      'size': 14,
                                      'color': 'black'})
ax.set_ylabel('Jabatan', fontdict = {'family': 'monospace',     # Setting font, ukuran font, dan warna font pada axis y
                                 'size': 14,
                                 'color': 'black'})
ax.set_title('Jabatan dengan Gaji Tertinggi\n', fontdict ={'family': 'monospace',       # Setting font, ukuran font, dan warna font pada judul grafik
                                                       'size': 18,
                                                       'color': 'black'} )
ax.set_xlim(150000,255000)                                                      # Limit axis x
plt.grid(axis = 'x', alpha = .3)
plt.legend()
```

Disini ditunjukkan 10 jabatan dengan pendapatan tertinggi. Chief Technology Officer serta CEO memiliki pendapatan tertinggi.

### 7. Apakah usia memiliki pengaruh terhadap gaji?

Selanjutnya akan ditinjau pendapatan seseorang menurut usia dari masing - masing pekerja. Dibuat grafik dengan bentuk scatter plot yang menunjukkan penyebaran data antara gaji dengan umur.

```python
# Pembuatan grafik scatter plot antara umur dengan gaji
fig, ax = plt.subplots()

ax.set_title('Gaji Karyawan Menurut Umur')
ax.set_xlabel('Umur')
ax.set_ylabel('Gaji')

ax.scatter(df['Age'],df['Salary'])
plt.show()
```

Dari grafik yang sudah dibuat, dapat dilihan bahwa semakin tua umur seseorang maka gaji nya semakin tinggi. Aspek ini juga dipengaruhi oleh pengalaman seseorang dalam bekerja, sehingga pengalaman bekerja memiliki pengaruh terhadap gaji seseorang, yang berarti umur dari pekerja tersebut kemungkinan lebih tua dari yang tidak memiliki pengalaman.

### 8. Jenis kelamin apa yang memiliki gaji paling tinggi?

Terakhir akan dilihat jenis kelamin mana yang memiliki gaji lebih tinggi, apakah jenis kelamin memiliki pengaruh yang signifikan dengan gaji yang dimiliki oleh seseorang.

```python
# Pembuatan groupby antara kolom gender dengan salary
df.groupby('Gender')['Salary'].mean().plot(kind = 'bar', figsize = (12,6))
```

Dari grafik yang sudah dibuat, jenis kelamin yang memiliki penghasilan lebih tinggi adalah jenis kelamin laki - laki. Namun tidak ada perbedaan yang signifikan antara kedua jenis kelamin, sehingga pendapatan seseorang tidak dipengaruhi oleh jenis kelamin seseorang tersebut.
